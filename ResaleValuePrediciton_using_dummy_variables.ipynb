```python
import pandas as pd
```

```python
df = pd.read_csv("carprices.csv")
```

```python
df
```

### column renaming

```python
df = df.rename(columns={"Sell Price($)":'SellPrice',"Age(yrs)":"Age"})
```

### creating dummy variables

```python
dummies_df = pd.get_dummies(df['Car Model'])
```

```python
dummies_df
```

### concatenating the files

```python
merged_df = pd.concat([df,dummies_df],axis='columns')
```

```python
merged_df * 1
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
df = merged_df.drop(['BMW X5'],axis='columns')
df = df * 1
```

```python
df = df.drop(['Car Model'],axis='columns')
```

```python
X = df.drop(['SellPrice'],axis=1)
X
```

```python
Y = df.SellPrice
```

```python
model.fit(X,Y)
```

### Predicting the the re-sale value  for a Mercedes-Benz that is 7 years old with 79,000 mileage.

```python
model.predict([[79000,7,0,1]])
```

### finding the accuracy of the model

```python
model.score(X,Y)
```
